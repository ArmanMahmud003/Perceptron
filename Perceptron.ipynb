```python
import numpy as np
import pandas as pd
class Perceptron():
  def __init__(self, eta=0.01, n_iter=10):
    self.weights=np.random.randn(2)*1e-4
    self.bias=np.random.randn(1)
    print(f"Weights: {self.weights}\n Bias: {self.bias}")
    self.eta = eta
    self.n_iter = n_iter
  def activationfunction(self,X):
    z=np.dot(X,self.weights)+self.bias
    return np.where(z>0,1,0)
  def fit(self, X, y):
    self.X=X
    self.y=y
    for i in range(self.n_iter):
      print(f"{i}th epoch")
      y_hat=self.activationfunction(self.X)
      print(f"predicted value: \n{y_hat}")
      error=self.y-y_hat
      print(f"error: \n{error}")
      if(error==0).all(): # Corrected line: Added .all() to resolve ValueError
        print("Training completed")
        break
      self.weights=self.weights+self.eta*np.dot(self.X.T,error)
      self.bias=self.bias+self.eta*np.sum(error)
      print(f"updated weights: \n{self.weights}")
      print("#############\n")
  def predict(self,X):
    return self.activationfunction(X)
```

```python
data = {"x1": [0,0,1,1], "x2": [0,1,0,1], "y": [0,1,1,1]}

AND = pd.DataFrame(data)
X = AND.drop("y", axis=1)
y = AND['y']
y.to_frame()
X
```

```python
model = Perceptron(eta = 0.5, n_iter=10)
model.fit(X,y)
```

```python
model.predict(X)
```
